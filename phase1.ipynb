```python
#Combine the DataFrames into a single DataFrame using the PatientPKHash column
def join_dataframes(dataframes):
    """
    This function takes in a list of DataFrames and combines them into a single DataFrame
    using the PatientPKHash column , keeping only rows with matching PatientPKHash values.

    Args:
        dataframes(list): A list of DataFrames to be combined.

    Returns:
        Pandas.DataFrame: The combined DataFrame.

    Raises:
        ValueError: If the any of the DataFrames do not have a PatientPKHash column.
    """
    #Check if the PatientPKHash column exists in all DataFrames and have the same name
    pkhash_columns = set([col for df in dataframes for col in df.columns if col=='PatientPKHash'])
    if len(pkhash_columns) !=1:
        raise ValueError('All DataFrames must have a PatientPKHash column with the same name.')

    #Perform iterative inner merges based on PatientPKHash
    merged_df = dataframes[0]
    for df in dataframes[1:]:
        merged_df = merged_df.merge(df, on='PatientPKHash', how='inner')
    
    return merged_df
```

```python
#Combine CT_patients.csv and CT_AdverseEvents.csv
df = join_dataframes([df_1, df_2])
df.shape
```

```python
def remove_rows_with_most_na(df, threshold=0.7):
    """
    This function removes rows with more than a certain percentage of missing values.

    Args:
        df(pandas.DataFrame): The DataFrame to be cleaned.
        threshold(float): The threshold value for percentage of missing values. Default is 0.7.

    Returns:
        Pandas.DataFrame: The cleaned DataFrame.
    """
    #Get the percentage of missing values in each column
    missing_ratios = df.isna().mean(axis=1)

    #Remove rows with more than the threshold percentage of missing values
    df = df[missing_ratios < threshold]

    return df
```

```python
#Clean df_3
df_3 = remove_rows_with_most_na(df_3, threshold=0.25)
df_3.shape
```

```python
#Clean df
df = remove_rows_with_most_na(df, threshold=0.25)
df.shape
```

```python
#merge df_3 and df
df = join_dataframes([df, df_3])
df.shape
```
